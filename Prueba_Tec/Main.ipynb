````python
# Extraer el año de la fecha
df['año'] = df['fecha'].dt.year

# Calcular la diferencia entre 'ventas' y 'modelo_actual'
df['diferencia'] = df['ventas'] - df['modelo_actual']

# Inicializar una lista para almacenar los resultados
resultados = []

# Bucle para 'prod_id' del 1 al 1000
for prod_id in range(1, 1001):
    # Bucle para año del 2013 al 2020
    for año in range(2013, 2021):
        # Filtrar el DataFrame para el 'prod_id' y año específicos
        filtro = (df['prod_id'] == prod_id) & (df['año'] == año)
        
        # Calcular la suma de la diferencia para el 'prod_id' y año específicos
        suma_diferencia = df[filtro]['diferencia'].sum()
        
        # Agregar el resultado a la lista
        resultados.append({'prod_id': prod_id, 'año': año, 'suma_diferencia': suma_diferencia})

# Convertir la lista de resultados a un DataFrame
resultado_df = pd.DataFrame(resultados)

# Mostrar el resultado
print(resultado_df)
````

## Regresión líneal múltiple

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

df = pd.read_csv('desafio_modelo_de_demanda.csv')

# Convertir la fecha a valores numéricos, ya que la regresión lineal tampoco puede usar fechas directamente
df['fecha'] = pd.to_datetime(df['fecha'])
df['anio'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month

# Preparar los datos para el modelo
X = df[['anio', 'mes', 'prod_id', 'precio_promedio']]  # Características
y = df['ventas']  # Variable objetivo

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inicializar y entrenar el modelo de regresión lineal múltiple
modelo_lr = LinearRegression()
modelo_lr.fit(X_train, y_train)

# Realizar predicciones con el conjunto de prueba
predicciones = modelo_lr.predict(X_test)

# Calcular el error cuadrático medio (MSE), la raíz del error cuadrático medio (RMSE) y el coeficiente de determinación (R^2)
mse = mean_squared_error(y_test, predicciones)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, predicciones)

print(f'R2: {r2}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
````

## Regresión polinomial de grado 3

````python
from sklearn.preprocessing import PolynomialFeatures
df['fecha'] = pd.to_datetime(df['fecha'])
df['anio'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month

# Preparar los datos para el modelo
X = df[['anio', 'mes', 'prod_id', 'precio_promedio']]  # Características
y = df['ventas']  # Variable objetivo

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Transformar las características a su forma polinomial de grado 3
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Inicializar y entrenar el modelo de regresión lineal con características polinomiales
modelo_poly = LinearRegression()
modelo_poly.fit(X_train_poly, y_train)

# Realizar predicciones con el conjunto de prueba
predicciones = modelo_poly.predict(X_test_poly)

# Calcular el error cuadrático medio (MSE) y el coeficiente de determinación (R^2)
mse = mean_squared_error(y_test, predicciones)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, predicciones)

print(f'R2: {r2}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
````

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import matplotlib.pyplot as plt


# Convertir la fecha a valores numéricos, como el año y el mes, ya que Random Forest no puede usar fechas directamente
df['fecha'] = pd.to_datetime(df['fecha'])
df['anio'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month

# Preparar los datos para el modelo
X = df[['anio', 'mes', 'prod_id', 'precio_promedio']]  # Características
y = df['ventas']  # Variable objetivo

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inicializar y entrenar el modelo de Random Forest
modelo_rf = RandomForestRegressor(n_estimators=800, random_state=42)
modelo_rf.fit(X_train, y_train)

# Realizar predicciones con el conjunto de prueba
predicciones = modelo_rf.predict(X_test)

# Calcular el error cuadrático medio (MSE) y la raíz del error cuadrático medio (RMSE)
r2 = r2_score(y_test, predicciones)
mse = mean_squared_error(y_test, predicciones)
rmse = np.sqrt(mse)
print(f'R2: {r2}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
````

````python
def predecir_ventas_2024(modelo_rf, precios_2024,inicio,fin):
    # Generar las fechas de los primeros días de cada mes en 2024
    fechas = pd.date_range(start=inicio, end=fin, freq='MS')
    
    # Inicializar un DataFrame para las predicciones con las columnas deseadas
    predicciones_df = pd.DataFrame(columns=['prod_id', 'fecha', 'ventas_prediccion'])
    
    for fecha in fechas:
        # Preparar el dataframe de predicción para este mes
        df_mes = precios_2024.copy()
        df_mes['anio'] = fecha.year
        df_mes['mes'] = fecha.month
        
        # Predecir las ventas
        predicciones = modelo_rf.predict(df_mes[['anio', 'mes', 'prod_id', 'precio_promedio']])
        
        # Crear un DataFrame temporal para las predicciones de este mes
        # Directamente al crear el DataFrame, evitando el SettingWithCopyWarning
        df_pred_mes = pd.DataFrame({
            'prod_id': df_mes['prod_id'].values,  # Usar .values para asegurar una copia de los datos
            'fecha': [fecha] * len(df_mes),  # Repetir la fecha para coincidir con el número de filas
            'ventas_prediccion': predicciones
        })
        
        # Concatenar el DataFrame temporal con el DataFrame de predicciones
        predicciones_df = pd.concat([predicciones_df, df_pred_mes], ignore_index=True)
    
    return predicciones_df

# Asumiendo que `modelo_rf` es tu modelo entrenado y `precios_2024` es el DataFrame con los precios promedios por producto
predicciones_2024 = predecir_ventas_2024(modelo_rf, precios_2024, inicio='2024-01-01',fin='2024-12-31')
print(predicciones_2024)
````

````python
df = predicciones_2024

# Filtrar el DataFrame por prod_id
prod_id_filtrado = df[df['prod_id'] == 1]

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(prod_id_filtrado['fecha'], prod_id_filtrado['ventas_prediccion'], marker='o', linestyle='-', color='b')
plt.title('Ventas Predicción por Fecha para prod_id = 1')
plt.xlabel('Fecha')
plt.ylabel('Ventas Predicción')
plt.grid(True)
plt.show()
````

````python
# Calcular el promedio de ventas predichas por prod_id si hay múltiples entradas por producto
ventas_promedio_por_id = df.groupby('prod_id')['ventas_prediccion'].mean().reset_index()

# Gráfica de barras
plt.figure(figsize=(10, 6))
plt.bar(ventas_promedio_por_id['prod_id'], ventas_promedio_por_id['ventas_prediccion'], color='orange')
plt.title('Promedio de Ventas Predicción por prod_id')
plt.xlabel('prod_id')
plt.ylabel('Ventas Predicción Promedio')
plt.xticks(ventas_promedio_por_id['prod_id'])
plt.grid(axis='y')
plt.show()
````

````python
plt.figure(figsize=(10, 6))
plt.hist(df['ventas_prediccion'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución de Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.show()
````
